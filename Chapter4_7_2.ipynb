# 4.7.2 [실습] 키와 몸무게에 따른 체형 군집화

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

## 데이터 만들기

```python
df = pd.DataFrame(columns=["height", "weight"])
df.loc[0] = [185, 60]
df.loc[1] = [180, 60]
df.loc[2] = [185, 70]
df.loc[3] = [165, 63]
df.loc[4] = [155, 68]
df.loc[5] = [170, 75]
df.loc[6] = [175, 80]
```

## 데이터 시각화

```python
sns.lmplot(x="height", y="weight", data=df, fit_reg=False, scatter_kws={"s": 200})
```

## k 평균 군집

```python
data_points = df.values
kmeans = KMeans(n_clusters=3).fit(data_points)

kmeans.cluster_centers_
df["cluster_id"] = kmeans.labels_
```

## k 평균 군집 시각화

```python
sns.lmplot(
    x="height",
    y="weight",
    data=df,
    fit_reg=False,
    scatter_kws={"s": 150},
    hue="cluster_id",
)
```
